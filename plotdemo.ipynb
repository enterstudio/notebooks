# Plot demo

First, we need to import some libraries

```python
%matplotlib inline
import numpy as np
import scipy as sp
import scipy.stats
import matplotlib.pyplot as plt
```

Allocate some values as x values.

```python
a = np.array(np.linspace(1,20,1000))
```

Let's try some plotting

```python
y = np.sin(a ** 2) / a
z = 1 / a
z2 = -1 / a
yz = np.vstack((y,z,z2))
print(np.shape(yz))

plt.plot(a, y, a, z, a, z2)
```

Pythagoras

$$r=\sqrt{x^2+y^2}$$

```python
rv = sp.stats.binom(10,0.5)
x = range(1,10)
rv.pmf(x)
```

日本語入れられるのかな？

The binomial distribution with 5 tries explains the shape of mass spectrum of DBDE (Decabromodiphenyl oxide), which is

$$
\begin{pmatrix}
n \\
k
\end{pmatrix}
=
{}_{n}C_k
=
\frac{n!}{k!(n-k)!}
$$

```python
plt.vlines(x, 0, rv.pmf(x), colors='k', linestyles='-', lw=1,
         label='frozen pmf')
```
