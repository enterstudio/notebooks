```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

# Round towards zero

```python
x = np.arange(-4,4,0.01)
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x,np.fix(x));
ax.set_xticks(np.arange(-4,5))
ax.set_yticks(np.arange(-4,5))
ax.grid()
ax.set_xlabel('x');
ax.set_ylabel('fix(x)');
```

# Floor

```python
x = np.arange(-4,4,0.01)
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x,np.floor(x));
ax.set_xticks(np.arange(-4,5))
ax.set_yticks(np.arange(-4,5))
ax.grid()
ax.set_xlabel('x');
ax.set_ylabel('floor(x)');
```

# Ceil

```python
x = np.arange(-4,4,0.01)
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x,np.ceil(x));
ax.set_xticks(np.arange(-4,5))
ax.set_yticks(np.arange(-4,5))
ax.grid()
ax.set_xlabel('x');
ax.set_ylabel('ceil(x)');
```

# Round

```python
x = np.arange(-4,4,0.01)
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x,np.round(x));
ax.set_xticks(np.arange(-4,5))
ax.set_yticks(np.arange(-4,5))
ax.grid()
ax.set_xlabel('x');
ax.set_ylabel('round(x)');
```

# Simulated round

```python
x = np.arange(-4,4,0.01)
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x,np.floor(x + .5));
ax.set_xticks(np.arange(-4,5))
ax.set_yticks(np.arange(-4,5))
ax.grid()
ax.set_xlabel('x');
ax.set_ylabel('floor(x + .5)');
```

# Precision of floating point numbers

Floating point numbers have variable precision over their value range.
The plot below illustrates the minimum representable value in that range.
(Note that the actual bottom limit is two bits lower than this log-log plot,
but unfortunately we cannot plot it with double precision numbers.)

It is worth pointing out that the precision increases (in other words, the
minimum representable value decreases) as the value approaches 0.

```python
x = 2. ** np.arange(-10,10,1)
y = np.repeat(x,2)[:-2]
x = np.repeat(x,2)[1:-1]
fig, ax = plt.subplots(figsize=(10,10))
ax.loglog(x, y * np.power(2., -512))
ax.grid(which='both')
ax.set_xlabel('Floating value')
ax.set_ylabel('Precision');
```
